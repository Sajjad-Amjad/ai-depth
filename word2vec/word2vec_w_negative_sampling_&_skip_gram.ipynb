```python
vocabulary = {word: i for i, word in enumerate(word_counts.keys())}
```

## Training Data

```python
def generate_training_data(words, vocabulary, window_size=2):
    data = []
    for i in range(window_size, len(words) - window_size):
        target = vocabulary[words[i]]
        context = [vocabulary[words[j]] for j in range(i - window_size, i + window_size + 1) if j != i]
        for context_word in context:
            data.append((target, context_word))
    return data
```

```python
training_data = generate_training_data(words, vocabulary, window_size=1)
training_data
```

```python
training_data = generate_training_data(words, vocabulary, window_size=2)
training_data
```

## Model Archtecture

```python
import torch
import torch.nn as nn
```

```python
class SkipGramModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim):
        super(SkipGramModel, self).__init__()
        self.target_embedding = nn.Embedding(vocab_size, embedding_dim)
        self.context_embedding = nn.Embedding(vocab_size, embedding_dim)

    def forward(self, target_word, context_word):
        target_vector = self.target_embedding(target_word)
        context_vector = self.context_embedding(context_word)
        score = torch.matmul(target_vector, context_vector.t())
        return score
```

## Negative Samples

```python
import numpy as np
```

```python
def negative_sampling(vocabulary, target_word, num_samples=5):
    negative_samples = []
    while len(negative_samples) < num_samples:
        negative_sample = np.random.choice(list(vocabulary.values()))
        if negative_sample != target_word:
            negative_samples.append(negative_sample)
    return negative_samples
```

## Loss Function

```python
def loss_function(score, label):
    return torch.nn.functional.binary_cross_entropy_with_logits(score, label)
```

## Training Loop

```python
model = SkipGramModel(vocab_size=len(vocabulary), embedding_dim=50)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(100):
    total_loss = 0
    for target, context in training_data:
        context_negatives = negative_sampling(vocabulary, target)
        context_words = torch.tensor([context] + context_negatives, dtype=torch.long)
        labels = torch.tensor([1.0] + [0.0] * len(context_negatives), dtype=torch.float)

        optimizer.zero_grad()
        scores = model(torch.tensor([target], dtype=torch.long), context_words)

        # Squeeze the scores tensor
        scores = scores.squeeze(0)
        loss = loss_function(scores, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    print(f'Epoch {epoch}, Loss: {total_loss:.4f}')
```

## Evaluation : Cosine Similarity

```python
def cosine_similarity(word1, word2):
    # Check if both words are in the vocabulary
    if word1 not in vocabulary or word2 not in vocabulary:
        raise KeyError(f"One or both words ('{word1}', '{word2}') not found in the vocabulary.")

    vec1 = model.target_embedding(torch.tensor([vocabulary[word1]], dtype=torch.long)).detach().numpy()
    vec2 = model.target_embedding(torch.tensor([vocabulary[word2]], dtype=torch.long)).detach().numpy()

    cos_sim = np.dot(vec1, vec2.T) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
    return cos_sim
```

## Evaluating : Analogy Task

```python
def analogy(word1, word2, word3):
    # Check if all words are in the vocabulary
    if word1 not in vocabulary or word2 not in vocabulary or word3 not in vocabulary:
        raise KeyError(f"One or more words ('{word1}', '{word2}', '{word3}') not found in the vocabulary.")

    vec = model.target_embedding(torch.tensor([vocabulary[word1]], dtype=torch.long)) - \
          model.target_embedding(torch.tensor([vocabulary[word2]], dtype=torch.long)) + \
          model.target_embedding(torch.tensor([vocabulary[word3]], dtype=torch.long))

    cos_similarities = torch.matmul(vec, model.target_embedding.weight.t()).detach().numpy()
    best_word_index = np.argmax(cos_similarities)
    best_word = [word for word, index in vocabulary.items() if index == best_word_index][0]

    return best_word
```

```python
# Test analogies
analogy_tests = [
    ("prince", "kingdom", "adventure"),
    ("galaxy", "stars", "cosmos"),
    ("relics", "treasure", "kingdom"),
    ("journey", "adventure", "creatures"),
    ("spaceship", "cosmos", "galaxy")
]
```

```python
for word1, word2, word3 in analogy_tests:
    try:
        result = analogy(word1, word2, word3)
        print(f"Analogy: '{word1}' - '{word2}' + '{word3}'  '{result}'")
    except KeyError as e:
        print(e)
```
